```julia
using Plots, LaTeXStrings, Optim

# define data
x = 1:10
y = [0.2,0.5,0.3,3.5,1.0,1.5,1.8,2.0,2.3,2.2]

# define fit functions
fit1(x0) = maximum(abs.(x0[1]*x .+ x0[2].-y))
fit2(x0) = sum(abs.(x0[1]*x .+ x0[2].-y))
fit3(x0) = sum(abs.(x0[1]*x .+ x0[2].-y).^2)

# fit data
x0 = [1.0, 1.0]

res1 = optimize(fit1, x0, NelderMead())
res2 = optimize(fit2, x0, NelderMead())
res3 = optimize(fit3, x0, NelderMead())

linear(res, x) = res.minimizer[1]*x .+ res.minimizer[2] # linear output

xf = 1:0.1:10
y1 = linear(res1, xf)
y2 = linear(res2, xf)
y3 = linear(res3, xf)

plot(x, y, seriestype = :scatter, label = "data")
plot!(xf, y1, label = L"E_{\infty}")
plot!(xf, y2, label = L"E_1")
plot!(xf, y3, label = L"E_2")
```
